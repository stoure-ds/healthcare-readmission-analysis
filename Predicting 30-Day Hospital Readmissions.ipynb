# Predictive modeling

```python
# Step 1 Encode categorical model

df_model = pd.get_dummies(df, drop_first=True)
df_model.head()
```

```python
# Step 2 Split the data

from sklearn.model_selection import train_test_split
X = df_model.drop('readmit_binary', axis=1)
y = df_model['readmit_binary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Step 3 Logistic Regression 

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model = LogisticRegression(max_iter=1000)
```

#### Explaination for max_iter=10000 
Logistic regression uses iterative optimization. With large datasets and many encoded features, the default iteration limit may be insufficient for convergence. Increasing max_iter ensures the model fully converges and produces stable coefficients.


```python
# fit the model

model.fit(X_train, y_train)
```

```python
# Predict the model

pred = model.predict(X_test)
print(classification_report(y_test, pred))
```

#### The model predicts:
every patient = NOT readmitted, it completely ignores class 1. That's why
- precision (class 1) = 0, recall (class 1) = 0. The model never predicted a single readmittion.
Why this happened:
- the dataset shows 
- Not readmitted: 90,409
- Readmitted: 11,357
That's ~89% of class 0, and ~11% of class 1.
This is called CLASS IMBALANCE
- the model learns: "if I always predict 0, I'll be right 89% of the time."


Initial logistic regression failed to detect readmission cases due to severe class imbalance, demonstrating the limitations of accuracy as a performance metric in healthcare risk prediction.

```python
# Fix the problem by forcing the model to care about minority cases by using Class Weighting.

model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train, y_train)
```

```python
# Make prediction

pred = model.predict(X_test)
print(classification_report(y_test, pred))
```

### Interpretation
A balanced logistic regression model significantly improved detection of 30-day readmissions, increasing recall from 0% to 43%. 

Although precision remains low due to class imbalance, prioritizing recall aligns with healthcare objectives where missing high-risk patients carries greater clinical and financial consequences than false alarms.

# ROC Curve

```python
# ROC Curve

from sklearn.metrics import roc_curve, auc

probs = model.predict_proba(X_test)[:,1]
fpr, tpr, _ = roc_curve(y_test, probs)
roc_auc = auc(fpr, tpr)
```

```python
# Plot the ROC curve

plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

Despite class balancing, the predictive model achieved an AUC of 0.51, indicating limited separability between readmitted and non-readmitted patients. This suggests that the available structured clinical features alone are insufficient to reliably predict readmission risk and that additional data sources (e.g., social determinants, follow-up care data, physician notes) may be required.

## Conclusion

The modeling results highlight a common challenge in healthcare analytics: structured administrative data alone may not capture the full complexity of patient risk. Future improvements could incorporate clinical text, social risk factors, and post-discharge follow-up indicators.
